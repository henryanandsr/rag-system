```python
df.columns = [
    'Perusahaan',
    'Produk',
    'Kalori',
    'Kalori dari Lemak',
    'Total Lemak',
    'Lemak Jenuh (g)',
    'Lemak tak Jenuh (g)',
    'Kolestrol (mg)',
    'Sodium (mg)',
    'Karbohidrat (g)',
    'Serat (g)',
    'Gula (g)',
    'Protein (g)',
    'WW Poin'
]
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
numerical_columns = [
    'Kalori', 'Kalori dari Lemak', 'Total Lemak', 'Lemak Jenuh (g)', 
    'Lemak tak Jenuh (g)', 'Kolestrol (mg)', 'Sodium (mg)', 'Karbohidrat (g)', 
    'Serat (g)', 'Gula (g)', 'Protein (g)', 'WW Poin'
]
for col in numerical_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
from sklearn.impute import SimpleImputer

numerical_imputer = SimpleImputer(strategy='mean')
df[numerical_columns] = numerical_imputer.fit_transform(df[numerical_columns])
```

```python
df.isnull().sum()
```

```python
df['Perusahaan'].value_counts()
```

```python
company_mapping = {
    "McDonald’s": "McDollops",
    "KFC": "KrispyFiestaChicken",
    "Burger King": "Raja Burger",
    "Taco Bell": "Lonceng Teko",
    "Wendy’s": "Berangin",
    "Pizza Hut": "Pizza Topi"
}
df["Perusahaan"] = df["Perusahaan"].replace(company_mapping)
```

```python
df.to_csv('../doc/fast-food.csv', index=False)
```
